- 小試身手: 把AQI, 硫化物, 還有站點的GPS座標印出來

````python
# 把程式填入這邊
for item in data:
    if item['County'] == "高雄市":
        print(item['County'], item['SiteName'], item['AQI'], item['SO2'], item['Longitude'], item['Latitude'])
````

````python
# 把程式填入這邊
for item in data:
    if item['County'] == "高雄市":
        #print(item['County'], item['SiteName'], item['AQI'], item['SO2'], item['Longitude'], item['Latitude'])
        print('[%s] %s: AQI=%3d, PM2.5=%2.2f, GPS=(%f, %1.3f)' 
              %(item['County'], item['SiteName'], int(item['AQI']), 
              float(item['SO2']), float(item['Longitude']), float(item['Latitude']) ) )
````

### 進一步找出 全台有超標的站
- AQI 是否超標，請參照以下的資訊
- ref: https://airtw.epa.gov.tw/CHT/Information/Standard/AirQualityIndicator.aspx

````python
# 把程式填入這邊

for item in data:
    if int(item['AQI']) >= 50:
        print('[%s] %s: AQI=%3d, SO2=%2.2f, GPS=(%f, %1.3f)' 
              %(item['County'], item['SiteName'], int(item['AQI']), 
              float(item['SO2']), float(item['Longitude']), float(item['Latitude']) ) )
````

### 在地圖上把所有的站畫出來

````python
import folium
import geocoder
````

````python
# 使用 geocoder 取得特定住址的 GPS 座標
location = geocoder.osm('台中市').latlng
````

````python
location
````

````python
m = folium.Map(location=location, zoom_start=15)
````

````python
print(data[0])
````

````python
for item in data:
    try:
        lat = float(item['Latitude'])
        lng = float(item['Longitude'])
        msg = 'AQI=%s, PM2.5=%s' %(item['AQI'], item['PM2.5'])
        folium.Marker([lat, lng], popup=msg).add_to(m)
    except Exception as e:
        print(e.args)    
````

### 畫出地圖，並且把超標的畫紅色、沒超標的畫綠色

````python
data[0]
````

````python
m = folium.Map(location=location, zoom_start=10)
````

````python
for item in data:
    try:
        lat = float(item['Latitude'])
        lng = float(item['Longitude'])
        aqi = int(item['AQI'])
        msg = '[%s] <br>AQI=%s <br>PM2.5=%s' %(item['SiteName'], item['AQI'], item['PM2.5'])
        if aqi>40:
            folium.Marker([lat, lng], popup=msg,
                         icon=folium.Icon(color='red')).add_to(m)
        else:
            folium.Marker([lat, lng], popup=msg,
                         icon=folium.Icon(color='green')).add_to(m)
    except Exception as e:
        print(e.args)    
````
