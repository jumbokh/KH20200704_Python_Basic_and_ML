### 抓sqlite資料
- 欄位: PublishTime, County, SiteName, AQI, PM10

```python
# 取出某段時間的資料
import sqlite3
def get_sqi_data_by_time(str_tm1):
    aqi_db = 'aqi_db.db'
    conn = sqlite3.connect(aqi_db)
    cursor = conn.cursor()
    # select last 5 rows
    res = cursor.execute("SELECT * FROM aqi WHERE PublishTime >= '%s' ORDER BY PublishTime ASC" %(str_tm1))
    rows = res.fetchall()
    
    cursor.close()
    conn.commit()
    conn.close()
    
    data_list = list(rows)
    return data_list
    
```

```python
get_sqi_data_by_time('2020-07-30')
```

```python
mydata = get_sqi_data_by_time('2020-07-28')
print(len(mydata))
```
