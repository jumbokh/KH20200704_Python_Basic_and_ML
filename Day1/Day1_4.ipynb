# Day1 - 3 : 公共腳踏車應用 - 進一步進化，連icon也可以改

## 新竹市公共自行車租賃系統(YouBike): https://data.gov.tw/dataset/67781

```python
import pandas as pd
```

```python
df = pd.read_csv("http://opendata.hccg.gov.tw/dataset/1f334249-9b55-4c42-aec1-5a8a8b5e07ca/resource/ccdcb6e2-58e8-46bf-adfe-9d31b4bf00ba/download/20191231090605186.csv")
```

```python
df.head()
```

```python
# 可以使用 `pip install folium` 安裝套件
# 使用 `pip install geocoder`
```

```python
import folium
import geocoder
```

```python
# 使用 geocoder 取得特定住址的 GPS 座標
location = geocoder.osm('新竹市').latlng
```

```python
location
```

```python
m = folium.Map(location=location, zoom_start=16)
```

```python
for i, row in df.iterrows():
    try:
        folium.Marker([float(row['緯度']), float(row['經度'])], popup=row['站點名稱'], 
                     icon=folium.Icon(color='green', prefix='fa', icon='fa-bicycle')).add_to(m)
    except Exception as e:
        print(e.args)
```

```python
m
```

- export出來的html就不會有中文顯示亂碼的問題

```python
m.save('新竹市bike.html')
```
